I was interested in an indicator that measures the CO2 emissions produced by Internal Combustion Engine (ICE) vehicles. Two indicator values that stood out were:
* 'CO2 emissions from liquid fuel consumption (kt)'
* 'CO2 emissions from transport (% of total fuel combustion)'

At this point, I needed to decide which of these two indicators most resembled CO2 emissions produced by ICE vehicles. 

The 'CO2 emissions from liquid fuel consumption (kt)' indicator potentially included emissions from ICE vehicles, however, it might have also included emissions from other sources. Because I wanted to focus on emissions from ICE vehicles, I decided to use the 'CO2 emissions from transport (% of total fuel combustion)' indicator. A potential issue with using this indicator, however, was that it reported a "% of total fuel combustion" rather a measure in **kt**. I decided that it was necessary to have a measure in **kt** to compare across different countries.

I performed a check for different indicator names that measured values in **kt**.

```python
kt_mask = data['IndicatorName'].str.contains('kt')
kt_stage = data[kt_mask]['IndicatorName'].unique().tolist()
print("No. of Indicator Name(s) containing \"kt\": ", len(kt_stage))
kt_stage
```

The fuel used in combustion can be solid, liquid, or gas\[1\]. *CO2 emissions from total fuel combustion* can, therefore, be calculated as the sum of the following measures:
* 'CO2 emissions from liquid fuel consumption (kt)'
* 'CO2 emissions from solid fuel consumption (kt)'
* 'CO2 emissions from gaseous fuel consumption (kt)'

To bring this information together and calculate emissions from total fuel combustion for one country, I created a new data frame. This process was later defined as a function in **Step 4** of this project.

```python
# Regular expressions to match the desired indicator names
liq_fuel_indicator = 'CO2 emissions from liquid.*kt\)'
sol_fuel_indicator = 'CO2 emissions from solid.*kt\)'
gas_fuel_indicator = 'CO2 emissions from gaseous.*kt\)'
transport_indicator = 'CO2 emissions from transport.*combustion\)'

# Create masks to identify rows from the world development indicators data that meet
#   certain conditions.
united_states_mask = data['CountryName'] == 'United States'
liq_fuel_mask = data['IndicatorName'].str.contains(liq_fuel_indicator)
sol_fuel_mask = data['IndicatorName'].str.contains(sol_fuel_indicator)
gas_fuel_mask = data['IndicatorName'].str.contains(gas_fuel_indicator)
transport_mask = data['IndicatorName'].str.contains(transport_indicator)

# Filter out data for CO2 emissions from liquid fuel consumption.
liq_fuel_stage = data[united_states_mask & liq_fuel_mask] \
    [['CountryName','IndicatorName','Year','Value']]
col_names = liq_fuel_stage.columns.values # Change column name "Value" to the indicator name.
co2_liq_name = liq_fuel_stage['IndicatorName'].iloc[0]
col_names[3] = co2_liq_name
liq_fuel_stage.columns = col_names
del liq_fuel_stage['IndicatorName'] # The 'IndicatorName' column is no longer needed

# Filter out data for CO2 emissions from solid fuel consumption.
sol_fuel_stage = data[united_states_mask & sol_fuel_mask][['IndicatorName','Year','Value']]
col_names = sol_fuel_stage.columns.values
co2_sol_name = sol_fuel_stage['IndicatorName'].iloc[0]
col_names[2] = co2_sol_name
sol_fuel_stage.columns = col_names
del sol_fuel_stage['IndicatorName']

# Filter out data for CO2 emissions from gaseous fuel consumption.
gas_fuel_stage = data[united_states_mask & gas_fuel_mask][['IndicatorName','Year','Value']]
col_names = gas_fuel_stage.columns.values
co2_gas_name = gas_fuel_stage['IndicatorName'].iloc[0]
col_names[2] = co2_gas_name
gas_fuel_stage.columns = col_names
del gas_fuel_stage['IndicatorName']

# Filter out data for CO2 emissions from transport as a percentage of total fuel combustion.
transport_stage = data[united_states_mask & transport_mask][['IndicatorName','Year','Value']]
col_names = transport_stage.columns.values
transport_name = transport_stage['IndicatorName'].iloc[0]
col_names[2] = transport_name
transport_stage.columns = col_names
del transport_stage['IndicatorName']

# Merge all of the stages on 'Year' to create the new data frame
united_states_transport = liq_fuel_stage.merge(sol_fuel_stage, on='Year', how='inner') \
    .merge(gas_fuel_stage, on='Year', how='inner') \
    .merge(transport_stage, on='Year', how='inner')

# Add a column for total CO2 emissions from combustion in kt
total_combustion_name = 'CO2 emissions from fuel combustion, total (kt)'
united_states_transport[total_combustion_name] = united_states_transport[co2_liq_name] \
    + united_states_transport[co2_sol_name] + united_states_transport[co2_gas_name]

# Add a column for CO2 emissions from transport in kt
kt_transport_name = 'CO2 emissions from transport (kt)'
united_states_transport[kt_transport_name] = (1/100) * united_states_transport[transport_name] \
    * united_states_transport[total_combustion_name]
```

## Write functions to generate a data frame containing CO2 emissions data for each country
As shown in **Step 2** of this project, the CO2 emissions data for a specific country can be collected from the WDI dataset and stored in a data frame. I wrote two functions to perform this task below.

```python
def generateCountryDF(country, data):
    """
    Uses a country name and the World Development Indicators(WDI) dataset to create a
    data frame consisting of information related to CO2 emissions from transport.
    Returns the final data frame.
    
    country: String object containing the country name
    data: Pandas data frame containing WDI dataset
    """
    liq_fuel_indicator = 'CO2 emissions from liquid.*kt\)'
    sol_fuel_indicator = 'CO2 emissions from solid.*kt\)'
    gas_fuel_indicator = 'CO2 emissions from gaseous.*kt\)'
    transport_indicator = 'CO2 emissions from transport.*combustion\)'
    pop_indicator = 'Population, total'

    country_mask = data['CountryName'] == country
    liq_fuel_mask = data['IndicatorName'].str.contains(liq_fuel_indicator)
    sol_fuel_mask = data['IndicatorName'].str.contains(sol_fuel_indicator)
    gas_fuel_mask = data['IndicatorName'].str.contains(gas_fuel_indicator)
    transport_mask = data['IndicatorName'].str.contains(transport_indicator)
    pop_mask = data['IndicatorName'] == pop_indicator

    liq_fuel_stage, co2_liq_name = generateStage(country_mask, liq_fuel_mask, data, \
                                                 ['CountryName','IndicatorName','Year','Value'])
    sol_fuel_stage, co2_sol_name = generateStage(country_mask, sol_fuel_mask, data, \
                                                ['IndicatorName','Year','Value'])
    gas_fuel_stage, co2_gas_name = generateStage(country_mask, gas_fuel_mask, data, \
                                                ['IndicatorName','Year','Value'])
    transport_stage, transport_name = generateStage(country_mask, transport_mask, data, \
                                                ['IndicatorName','Year','Value'])
    pop_stage, pop_name = generateStage(country_mask, pop_mask, data, \
                                                ['IndicatorName','Year','Value'])

    country_transport = liq_fuel_stage.merge(sol_fuel_stage, on='Year', how='inner') \
        .merge(gas_fuel_stage, on='Year', how='inner') \
        .merge(transport_stage, on='Year', how='inner') \
        .merge(pop_stage, on='Year', how='inner')

    total_combustion_name = 'CO2 emissions from fuel combustion, total (kt)'
    country_transport[total_combustion_name] = country_transport[co2_liq_name] \
        + country_transport[co2_sol_name] + country_transport[co2_gas_name]

    kt_transport_name = 'CO2 emissions from transport (kt)'
    country_transport[kt_transport_name] = (1/100) * country_transport[transport_name] \
        * country_transport[total_combustion_name]
    
    mtpc_transport_name = 'CO2 emissions from transport (metric tons per capita)'
    country_transport[mtpc_transport_name] = 1000 * country_transport[kt_transport_name] \
        / country_transport[pop_name]
    
    return country_transport
```

```python
def generateStage(country_mask, indicator_mask, data, col_names):
    """
    Creates a data frame of filtered values and columns from the WDI dataset. Masks
    for the country and desired indicator value are used for the filter. Returns
    the filtered data frame and name of the indicator value.
    
    country_mask: Pandas series of boolean values to indicate if a row is attributed to a
        specific country. Series length is same as the number of columns in "data".
    indicator_mask: Pandas series of boolean values to indicate if a row is attributed to
        a specific indicator (e.x. "CO2 emissions (kt)"). Series length is same as the
        number of columns in "data".
    data: Pandas data frame containing World Development Indicators dataset
    col_names: List of column names to pull from "data".
    """
    indicator_stage = data[country_mask & indicator_mask][col_names]
    indicator_name = indicator_stage['IndicatorName'].iloc[0]
    col_names[len(col_names) - 1] = indicator_name
    indicator_stage.columns = col_names
    del indicator_stage['IndicatorName']
    return indicator_stage, indicator_name
```

## Generate graphs to compare the CO2 emissions from transport between different countries
### CO2 Emissions Per Capita Over Time

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize = (12, 6))
for country_transport in countries_transport:
    plt.plot(country_transport['Year'].values, \
             country_transport['CO2 emissions from transport (metric tons per capita)'].values, \
             label = country_transport['CountryName'].iloc[0] + " (" \
             + str(country_transport['Year'].min()) + "-" \
             + str(country_transport['Year'].max()) + ")")
    
plt.xlabel('Year', fontsize = 12)
plt.ylabel('CO2 emissions from transport (metric tons per capita)', fontsize = 12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.title('CO2 Emissions from Transport per Capita')

plt.show()
```

Because there was unavailable data for certain countries within this time period, I decided it was better to truncate the graph to the years in which data was available for all countries.

```python
import numpy as np

plt.figure(figsize = (12, 6))
for country_transport in countries_transport:
    min_yr_mask = country_transport['Year'] >= 1971
    max_yr_mask = country_transport['Year'] <= 2011
    trunc_transport = country_transport[min_yr_mask & max_yr_mask]
    plt.plot(trunc_transport['Year'].values, \
             trunc_transport['CO2 emissions from transport (metric tons per capita)'].values, \
             label = country_transport['CountryName'].iloc[0])
    
plt.xlabel('Year', fontsize = 12)
plt.ylabel('CO2 emissions from transport (metric tons per capita)', fontsize = 12)
plt.xticks(np.arange(1971, 2012, 4))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('CO2 Emissions from Transport per Capita')

# plt.savefig('CO2_Emissions_Transport_PC.png')
plt.show()
```

### Total CO2 Emissions Bar Graphs
The bar graph below shows CO2 emissions from transport for only the year 2011.

```python
country_emissions = []
non_us_emissions = 0
for country_transport in countries_transport:
    yr_mask = country_transport['Year'] == 2011
    trunc_transport = country_transport[yr_mask]
    country_emissions.append(trunc_transport['CO2 emissions from transport (kt)'].iloc[0]
                             / (10**6))
    if trunc_transport['CountryName'].iloc[0] != 'United States':
        non_us_emissions += trunc_transport['CO2 emissions from transport (kt)'].iloc[0] \
            / (10**6)

temp_list = []
for i in range(len(country_emissions)):
    temp_list.append(tuple([country_emissions[i], countries[i]]))
def sortFunc(e):
    return e[0]
temp_list.sort(key = sortFunc, reverse = True)
temp_countries = []
for j in range(len(temp_list)):
    country_emissions[j] = temp_list[j][0]
    temp_countries.append(temp_list[j][1])
country_emissions.append(non_us_emissions)
temp_countries.append('Non U.S. Combined')
country_emissions = tuple(country_emissions)
temp_countries = tuple(temp_countries)


ind = np.arange(len(countries_transport) + 1)
width = 0.3
colors = ['blue'] * len(countries_transport)
colors[0] = 'crimson'
colors.append('green')

plt.figure(figsize = (6, 6))
plt.bar(ind, country_emissions, width, color = colors)

plt.ylabel('CO2 Emissions from Transport (Millions of kt)', fontsize = 12)
plt.title('2011 CO2 Emissions from Transport')
plt.xticks(ind, temp_countries, rotation = 60, fontsize = 12)

# plt.savefig('CO2_Emissions_Transport_2011.png')
plt.show()
```

The bar graph below shows the cumulative CO2 emissions from transport for each country from the years 1971 to 2011.

```python
country_sums = []
non_us_emissions = 0
for country_transport in countries_transport:
    min_yr_mask = country_transport['Year'] >= 1971
    max_yr_mask = country_transport['Year'] <= 2011
    trunc_transport = country_transport[min_yr_mask & max_yr_mask]
    country_sums.append(trunc_transport['CO2 emissions from transport (kt)'].sum()
                        / (10**6))
    if trunc_transport['CountryName'].iloc[0] != 'United States':
        non_us_emissions += trunc_transport['CO2 emissions from transport (kt)'].sum() \
            / (10**6)

temp_list = []
for i in range(len(country_sums)):
    temp_list.append(tuple([country_sums[i], countries[i]]))
temp_list.sort(key = sortFunc, reverse = True)
temp_countries = []
for j in range(len(temp_list)):
    country_sums[j] = temp_list[j][0]
    temp_countries.append(temp_list[j][1])
country_sums.append(non_us_emissions)
temp_countries.append('Non U.S. Combined')
country_sums = tuple(country_sums)
temp_countries = tuple(temp_countries)

ind = np.arange(len(countries_transport) + 1)
width = 0.3
colors = ['blue'] * len(countries_transport)
colors[0] = 'crimson'
colors.append('green')

plt.figure(figsize = (6, 6))
plt.bar(ind, country_sums, width, color = colors)

plt.ylabel('CO2 Emissions from Transport (Millions of kt)', fontsize = 12)
plt.title('Cumulative CO2 Emissions from Transport (1971 to 2011)')
plt.xticks(ind, temp_countries, rotation = 60, fontsize = 12)

# plt.savefig('CO2_Emissions_Transport_Cumulative.png')
plt.show()
```
